```python
from datetime import datetime

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from poliastro import ephem, iod, twobody
from poliastro import plotting
from poliastro.constants import k_Sun
```

```python
date_launch = datetime(2011, 11, 26, 15, 2)
date_arrival = datetime(2012, 8, 6, 5, 17)
tof = (date_arrival - date_launch).total_seconds()
```

```python
coe0_earth = ephem.mean_elements(date_launch, ephem.EARTH)
coef_earth = ephem.mean_elements(date_arrival, ephem.EARTH)
coe0_mars = ephem.mean_elements(date_launch, ephem.MARS)
coef_mars = ephem.mean_elements(date_arrival, ephem.MARS)

r0_earth, v0_earth = twobody.coe2rv(k_Sun, *coe0_earth)
rf_earth, vf_earth = twobody.coe2rv(k_Sun, *coef_earth)
r0_mars, v0_mars = twobody.coe2rv(k_Sun, *coe0_mars)
rf_mars, vf_mars = twobody.coe2rv(k_Sun, *coef_mars)
```

```python
r0 = r0_earth
rf = rf_mars
va, vb = iod.lambert(k_Sun, r0, rf, tof)
coe0_trans = twobody.rv2coe(k_Sun, r0, va)
coef_trans = twobody.rv2coe(k_Sun, rf, vb)
```

```python
nu0_earth = coe0_earth[-1]
nuf_earth = coef_earth[-1]
nu0_mars = coe0_mars[-1]
nuf_mars = coef_mars[-1]
nu0_trans = coe0_trans[-1]
nuf_trans = coef_trans[-1]
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

color_earth0 = '#3d4cd5'
color_earthf = '#525fd5'
color_mars0 = '#ec3941'
color_marsf = '#ec1f28'
color_sun = '#ffcc00'
color_orbit = '#888888'
color_trans = '#444444'

plotting.plot_orbit(ax, k_Sun, *coe0_earth[:-1], nu_lims=(nuf_earth, nu0_earth), ls='--', c=color_orbit)
plotting.plot_orbit(ax, k_Sun, *coef_mars[:-1], nu_lims=(nuf_mars, nu0_mars), ls='--', c=color_orbit)

plotting.plot_orbit(ax, k_Sun, *coe0_earth[:-1], nu_lims=(nu0_earth, nuf_earth), c=color_earth0)
plotting.plot_orbit(ax, k_Sun, *coe0_mars[:-1], nu_lims=(nu0_mars, nuf_mars), c=color_marsf)

plotting.plot_orbit(ax, k_Sun, *coe_trans[:-1], nu_lims=(nu0_trans, nuf_trans), c=color_trans)

ax.plot([0], [0], [0], 'o', color=color_sun, ms=16, mew=0)

ax.plot([r0[0]], [r0[1]], [r0[2]], 'o', c=color_earth0, ms=8, mew=0)
ax.plot([rf_earth[0]], [rf_earth[1]], [rf_earth[2]], 'o', c=color_earthf, ms=8, mew=0)

ax.plot([r0_mars[0]], [r0_mars[1]], [r0_mars[2]], 'o', c=color_mars0, ms=8, mew=0)
ax.plot([rf[0]], [rf[1]], [rf[2]], 'o', c=color_marsf, ms=8, mew=0)

ax.text(-0.75e8, -3.5e8, -1e8, "MSL mission:\nfrom Earth to Mars", size=20, backgroundcolor='w', ha='center', va='center', bbox={"pad": 30, "lw": 0})
ax.text(r0[0] * 1.15, r0[1] * 1.25, r0[2] * 10, "Earth at launch\n(26 Nov)", ha="left", va="bottom")
ax.text(rf[0] * 0.7, rf[1] * 1.1, rf[2], "Mars at arrival\n(6 Ago)", ha="left", va="top")
ax.text(-1.9e8, 8e7, 0, "Transfer\norbit", ha="right", va="center")

ax.set_xlim(-3e8, 3e8)
ax.set_ylim(-3e8, 3e8)
ax.set_zlim(-3e8, 3e8)

ax.view_init(80, 260)
fig.savefig("trans_80_260.png", bbox_inches='tight')
ax.view_init(30, 260)
fig.savefig("trans_30_260.png", bbox_inches='tight')
```
